# 2.1 Univariate regression on analytical functions

$$
f_1(x) = x\cdot \sin(x) + 2x
$$
$$
f_2(x) = 10 \sin(x) + x^2
$$
$$
f_3(x) = sign(x)(x^2+300) + 20\sin(x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split

tr = 20
n_samples = 100
X = np.linspace(-tr, tr, n_samples)
y_1 = X*np.sin(X) + 2*X
y_2 = 10*np.sin(X) + X**2
y_3 = np.sign(X)*(X**2 + 300) + 20*np.sin(X)
X_train, X_test, y_train, y_test = train_test_split(
                        X, y_1, train_size=0.7, 
                        random_state=42, shuffle=True
                        )
y_test = y_test[X_test.argsort()]
X_test.sort()
```

```python
plt.figure(figsize=(12,10))
plt.plot(X,y_1, label='f1(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Linear

```python
plt.figure(figsize=(12,10))
plt.plot(X,y_2, label='f2(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Polinomial $x^2$

```python
plt.figure(figsize=(12,10))
plt.plot(X,y_3, label='f3(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Polinomial grade > 2 (odd)

with polynomial regression we can try RIDGE (tends to lower uniformly all the coefficients) or LASSO (tends to assign the value 0 to some 
coefficients, feature selection)

```python
X_train, X_test, y_train, y_test = train_test_split(
                        X, y_1, train_size=0.7, 
                        random_state=42, shuffle=True
                        )
y_test = y_test[X_test.argsort()]
X_test.sort()

plt.figure(figsize=(12,10))
plt.scatter(X_test,y_test, label='y_true')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
from sklearn import linear_model

np.random.seed(0)

X_train, X_test, y_train, y_test = train_test_split(
                        X, y_1, train_size=0.7, 
                        random_state=42, shuffle=True
                        )
y_test = y_test[X_test.argsort()]
X_test.sort()

reg = linear_model.LinearRegression()
reg.fit(X_train.reshape(-1,1), y_train)
predictions = reg.predict(X_test.reshape(-1,1))

plt.figure(figsize=(12,10))
plt.scatter(X_test,y_test, label='y_true')
plt.plot(X_test, predictions, label='predictions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

from sklearn.metrics import mean_squared_error, r2_score
print(f'mean squared error: {mean_squared_error(y_test, predictions):.2f}')
print(f'r2_score: {r2_score(y_test, predictions):.2f}')
```
